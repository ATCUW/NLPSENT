```python
API_KEY = 'HGB2PFUGA9C14QWT'
BASE_LINK = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=TICKER&apikey=API_KEY'
```

```python
def sp500_tickers():
    resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    soup = bs.BeautifulSoup(resp.text, 'lxml')
    table = soup.find('table', {'class': 'wikitable sortable'})
    tickers = []
    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        tickers.append(ticker.strip())
        
    with open("sp500tickers.pickle","wb") as f:
        pickle.dump(tickers,f)
        
    return tickers
```

```python
tickers = sp500_tickers()
prices = []
```

```python
for ticker in tickers:
    resp = requests.get(BASE_LINK.replace('TICKER', ticker, 1).replace('API_KEY', API_KEY, 1))
    prices.append(resp.text)
    print(ticker + ' successful')
    time.sleep(12)
```
